## Integer and Conversions

* val: 25 type: <class 'int'> int: 25 float: 25.0 str: 25 bool: True
* val: -25 type: <class 'int'> int: -25 float: -25.0 str: -25 bool: True
* val: 0 type: <class 'int'> int: 0 float: 0.0 str: 0 **bool: False**
## Float and Conversions
* val: 25.0 type: <class 'float'> int: 25 float: 25.0 str: 25.0 bool: True
* val: -25.0 type: <class 'float'> int: -25 float: -25.0 str: -25.0 **bool: True**
* val: 25.97856 type: <class 'float'> int: 25 float: 25.97856 str: 25.97856 bool: True
* int('-25.97856') causes **Value Error**
* val: -25.97856 type: <class 'float'> int: -25 float: -25.97856 str: -25.97856 bool: True
* val: 0.0 type: <class 'float'> int: 0 float: 0.0 str: 0.0 bool: False
## String and Conversions
* val: 25 type: <class 'str'> int: 25 float: 25.0 str: 25 bool: True
* val: 0 type: <class 'str'> int: 0 float: 0.0 str: 0 **bool: True**
* val:  type: <class 'str'> int: float: str:  **bool: False**
* int(''), float('') causes **Value Error**
* val: -25 type: <class 'str'> int: -25 float: -25.0 str: -25 bool: True

* val: -25.97856 type: <class 'str'> int: float: -25.97856 str: -25.97856 bool: True
* int('sam'), float('sam') **causes Value Error**
* val: sam type: <class 'str'> int: float: str: sam bool: True

## Bool and Conversions
* val: True type: <class 'bool'> int: 1 float: 1.0 str: True bool: True
* val: False type: <class 'bool'> int: 0 float: 0.0 str: False bool: False

## Operations
* Division results in float
* modulo division results in Integer
* Equality check and other compariosn operators result in boolean

# Range function

```python
# Generate code below for Range function  and its usage in loops in Python showing all operations with explanation as comments
list1 = [1, 2, 3, 4, 5]
list2 = ['a', 'b', 'c']
range1 = range(5)  # Creates a range object from 0 to 4
print("list1:", list1)
print("list2:", list2)
print("range1:", list(range1))  # Converting range object to list for display
# Using range in for loop
print("Using range in for loop:")
for i in range(len(list1)):
    print(f"Index {i}, Element {list1[i]}")
# Using range with start, stop, step
print("Using range with start, stop, step:")
for i in range(1, 10, 2):  # From 1 to 9 with step of 2
    print(i)    
# Using range in reverse
print("Using range in reverse:")
for i in range(10, 0, -1):  # From 10 to 1
    print(i)
print("All range function operations executed successfully.")
print("\nRange Function and Its Usage in Loops in Python:")
# Generate code below for Range function and its difference between python 2 and python 3 showing all operations with explanation as comments
print("In Python 2, range() returns a list, while in Python 3, range() returns a range object which is more memory efficient.")
# Example in Python 2 (commented out as this code runs in Python 3)
# print("Python 2 range(5):", range(5))  # This would return [0, 1, 2, 3, 4]
# Example in Python 3
print("Python 3 range(5):", range(5))  # This returns a range object
print("Converted to list:", list(range(5)))  # Converting range object to list  for display
# Using range in for loop in Python 3
print("Using range in for loop in Python 3:")
for i in range(5):
    print(i)
print("All range function differences between Python 2 and Python 3 executed successfully.")
```

# Loops

```python
# Generate code below for Loops in Python showing all operations with explanation as comments
print("\nLoops in Python:")
# Example list
example_list = [1, 2, 3, 4, 5]
print("example_list:", example_list)
# for loop
print("Using for loop:")
for item in example_list:
    print(item)
# while loop
print("Using while loop:")
count = 0
while count < len(example_list):
    print(example_list[count])
    count += 1
# Using break statement
print("Using break statement in for loop:")
for item in example_list:
    if item == 3:
        print("Breaking the loop at item:", item)
        break
    print(item) 
# Using continue statement
print("Using continue statement in for loop:")
for item in example_list:
    if item == 3:
        print("Continuing the loop at item:", item)
        continue
    print(item)
print("All loop operations executed successfully.")
# Generate code below for Loops using string data type in Python showing all operations with explanation as comments
print("\nLoops with String Data Type in Python:")
# Example string
example_string = "Hello"
print("example_string:", example_string)
# for loop
print("Using for loop:")
for char in example_string:
    print(char)
# while loop
print("Using while loop:")
index = 0
while index < len(example_string):
    print(example_string[index])
    index += 1
# Using break statement
print("Using break statement in for loop:")
for char in example_string:
    if char == 'o':
        print("Breaking the loop at char:", char)
        break
    print(char)
# Using continue statement
print("Using continue statement in for loop:")
for char in example_string:
    if char == 'e':
        print("Continuing the loop at char:", char)
        continue
    print(char)
print("All loop operations with string data type executed successfully.")
```

# Functions

```python
# Generate code below for functions in Python
print("\nFunctions in Python:")
# Defining a simple function
def greet(name):
    return f"Hello, {name}!"
# Calling the function
print("greet('Alice'):", greet("Alice"))
# Function with default parameter
def power(base, exponent=2):
    return base ** exponent
print("power(3):", power(3))
print("power(2, 3):", power(2, 3))  
# Function with variable-length arguments
def sum_all(*args):
    return sum(args)
print("sum_all(1, 2, 3):", sum_all(1, 2, 3))
print("sum_all(4, 5, 6, 7, 8):", sum_all(4, 5, 6, 7, 8))
# Function with keyword arguments
def introduce(name, age):
    return f"My name is {name} and I am {age} years old."
print("introduce(name='Bob', age=30):", introduce(name="Bob", age=30))  
print("introduce(age=25, name='Charlie'):", introduce(age=25, name="Charlie"))
# Nested function
def outer_function(x):
    def inner_function(y):
        return y * 2
    return inner_function(x) + 3 
print("outer_function(5):", outer_function(5))
# function without return statement
def print_message(message):
    print("Message:", message)
    print_message("This is a test message.")
print("All function operations executed successfully.")
print("\nString Methods:")
s4 = "12345"
s5 = "HelloWorld"
s6 = "hello"
s7 = "WORLD"
print("s4.isalpha():", s4.isalpha())
print("s4.isalnum():", s4.isalnum())
print("s4.isdigit():", s4.isdigit())
print("s5.isalpha():", s5.isalpha())
print("s5.isalnum():", s5.isalnum())
print("s5.isdigit():", s5.isdigit())
print("s5.lower():", s5.lower())
print("s5.upper():", s5.upper())
print("s6.islower():", s6.islower())
print("s7.isupper():", s7.isupper())
print("s5.startswith('Hello'):", s5.startswith('Hello'))
print("s5.endswith('World'):", s5.endswith('World'))
name = "John"
age = 30
pi = 3.14159
print("My name is {}, age is {}, pi approx = {:.2f}".format(name, age, pi))
print("All string methods executed successfully.")


```

# Exception handling

```python
# Generate code below for exception handling in Python showing all operations with explanation as comments
print("\nException Handling in Python:")
# Example function that may raise an exception
def divide(a, b):
    return a / b
# Using try-except block
try:
    result = divide(10, 0)
    print("Result:", result) 
except ZeroDivisionError as e:
    print("Caught a division by zero error:", e)
# Using try-except-else block
try:
    result = divide(10, 2)
    print("Result:", result)
except ZeroDivisionError as e:
    print("Caught a division by zero error:", e)
else:
    print("Division successful, result is:", result)
# Using try-except-finally block
try:
    result = divide(10, 0)
    print("Result:", result)
except ZeroDivisionError as e:
    print("Caught a division by zero error:", e)
finally:
    print("Execution of try-except-finally block is complete.")
# Raising an exception
def check_positive(number):
    if number < 0:
        raise ValueError("Number must be positive")
    return True
try:
    check_positive(-5)
except ValueError as e:
    print("Caught a ValueError:", e)
print("All exception handling operations executed successfully.")

```

# Classes and objects

```python
#Generate code below for class and object in Python showing all operations with explanation as comments
print("\nClass and Object in Python:")
# Defining a class
class Person:
    def __init__(self, name, age):
        self.name = name  # Instance variable for name
        self.age = age    # Instance variable for age
    def greet(self):
        return f"Hello, my name is {self.name} and I am {self.age} years old."
# Creating an object (instance) of the class
person1 = Person("Alice", 25)
print("person1:", person1)
# Accessing object attributes
print("person1.name:", person1.name)
print("person1.age:", person1.age)
# Calling a method of the object
print("person1.greet():", person1.greet())
# Creating another object
person2 = Person("Bob", 30)
print("person2:", person2)
print("person2.greet():", person2.greet())
# Modifying object attributes
person1.age = 26
print("Modified person1.age:", person1.age)
print("All class and object operations executed successfully.")
    
```

```python
# Generate code below for file handling in Python showing all operations with explanation as comments
print("\nFile Handling in Python:")
# Writing to a file
with open('example.txt', 'w') as file:
    file.write("Hello, \n  this is a \n test file.\n")
    file.write("This file is used for demonstrating file handling in Python.\n")
print("File 'example.txt' created and written successfully.")
# Reading from a file
with open('example.txt', 'r') as file:
    content = file.read()
    print("Content of 'example.txt':")
    print(content)
# Appending to a file
with open('example.txt', 'a') as file:
    file.write("This line is appended to the file.\n")
print("Line appended to 'example.txt' successfully.")
# Reading the updated file
with open('example.txt', 'r') as file:
    updated_content = file.read()
    print("Updated content of 'example.txt':")
    print(updated_content)
# Using try-except for file handling
try:
    with open('non_existent_file.txt', 'r') as file:
        data = file.read()
except FileNotFoundError as e:
    print("Caught a FileNotFoundError:", e)
print("All file handling operations executed successfully.")
# generate code below for readlines() method in Python showing all operations with explanation as comments
print("\nUsing readlines() Method in Python:")
# Reading lines from a file using readlines()
with open('example.txt', 'r') as file:
    lines = file.readlines()  # This will read all lines and return a list of lines
    print("Lines read from 'example.txt':")
    for line in lines:
        print(line.strip())  # Using strip() to remove any extra whitespace
with open('example.txt', 'r') as file:
    line = file.readline(20)#   This will read the first 20 characters from the file until it encounters a newline character or reaches the end of the file
    print("Lines read from 'example.txt' using readline(20):", line)
    lines = file.read(20)# This will read the next 20 characters from the file until it reaches the end of the file
    print("Next 20 characters read from 'example.txt' using read(20):", lines)
with open('example.txt', 'r') as file:
    line = file.readline()  # This will read the next line from the file
    print("type of readline():", type(line))
    print('line[0]',line[0])
with open('example.txt', 'r') as file:
    line = file.read()  # This will read the entire file
    print("type of read():", type(line))
    print('line[0]',line[0])    #  This will print the first character of the entire file content
with open('example.txt', 'r') as file:
    lines = file.readlines()  # This will read all lines and return a list of lines
    print("type of readlines():", type(lines))
    print('lines[0]',lines[0])    # This will print the first line of the file
with open('example.txt', 'r') as file:
    lines = file.read(5)  # This will read the first 5 characters from the file
    print("type of read(5):", type(lines))
print("All readlines() method operations executed successfully.")   
```

# Pandas

```python
# install pandas library in Jupyter
%pip install pandas
#Generate code below for Pandas in Python showing all operations with explanation as comments
print("\nPandas in Python:")    
import pandas as pd
# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}   
df = pd.DataFrame(data)
print("DataFrame created from dictionary:")
print(df)
# Accessing DataFrame columns
print("Accessing 'Name' column:")
print(df['Name'])
# Accessing DataFrame rows using loc
print("Accessing row with index 1 using loc:")
print(df.loc[1])
# Accessing DataFrame rows using iloc
print("Accessing row with index 1 using iloc:")
print(df.iloc[1])
# Adding a new column to the DataFrame
df['Occupation'] = ['Engineer', 'Doctor', 'Artist']
print("DataFrame after adding 'Occupation' column:")
print(df)
# Filtering the DataFrame
filtered_df = df[df['Age'] > 28]
print("Filtered DataFrame with Age > 28:")
print(filtered_df)
# Grouping the DataFrame by 'City' and calculating the mean age
grouped_df = df.groupby('City')['Age'].mean()
print("Mean age grouped by City:")
print(grouped_df)
# Sorting the DataFrame by 'Age'
sorted_df = df.sort_values(by='Age')
print("DataFrame sorted by Age:")
print(sorted_df)
# Saving the DataFrame to a CSV file
df.to_csv('example.csv', index=False)
print("DataFrame saved to 'example.csv' successfully.")
# Reading the DataFrame from a CSV file
df_from_csv = pd.read_csv('example.csv')
print("DataFrame read from 'example.csv':")
print(df_from_csv)
#Slicing the DataFrame
sliced_df = df.iloc[0:2]  # This will slice the first two rows of the DataFrame
print("Sliced DataFrame (first two rows):")
print(sliced_df)
#Slicing the DataFrame with steps
sliced_df_steps = df.iloc[0:3:2]  # This will slice the first three rows with a step of 2 (i.e., it will take the 0th and 2nd rows)
print("Sliced DataFrame with steps (0th and 2nd rows):")
print(sliced_df_steps)
print("All Pandas operations executed successfully.")
```

# Numpy

```python
# Generate code below for Numpy in Python showing all operations with explanation as comments
print("\nNumpy in Python:")
import numpy as np
# Creating a NumPy array from a list
array1 = np.array([1, 2, 3, 4, 5])
print("NumPy array created from list:")
print(array1)
# Creating a 2D NumPy array
array2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D NumPy array:")
print(array2)
# Accessing elements in a NumPy array
print("Accessing element at index 0 in array1:", array1[0])
print("Accessing element at row 1, column 2 in array2:", array2[1, 2])
# Performing element-wise operations
array3 = array1 * 2
print("Element-wise multiplication of array1 by 2:")
print(array3)
array4 = array1 + array3
print("Element-wise addition of array1 and array3:")
print(array4)
# Calculating the mean and standard deviation of a NumPy array
mean_array1 = np.mean(array1)
std_array1 = np.std(array1)
print("Mean of array1:", mean_array1)
print("Standard deviation of array1:", std_array1)
# Reshaping a NumPy array
reshaped_array = array2.reshape(3, 2)
print("Reshaped array2 to shape (3, 2):")
print(reshaped_array)
# Saving a NumPy array to a file
np.save('array1.npy', array1)
print("NumPy array saved to 'array1.npy' successfully.")
# Loading a NumPy array from a file
loaded_array = np.load('array1.npy')
print("NumPy array loaded from 'array1.npy':")
print(loaded_array)
# Slicing a NumPy array
sliced_array = array1[1:4]  # This will slice the elements from index 1 to 3 (4 is exclusive)
print("Sliced array1 (elements from index 1 to 3):")
print(sliced_array)
# Slicing a 2D NumPy array
sliced_array2 = array2[0:2, 1:3]  # This will slice the first two rows and columns 1 to 2 (3 is exclusive)
print("Sliced array2 (first two rows and columns 1 to 2):")
print(sliced_array2)
print("All Numpy operations executed successfully.")
```

```python
#install requests library in Jupyter
%pip install requests
# Generate code below for Requests in Python showing all operations with explanation as comments
print("\nRequests in Python:")
import requests
# Making a GET request to a URL
response = requests.get('https://jsonplaceholder.typicode.com/posts/1')
print("GET request to 'https://jsonplaceholder.typicode.com/posts/1' made successfully.")
print("Status code of the response:", response.status_code)
print("Content of the response:")
print(response.text)
# Making a POST request to a URL
post_data = {
    'title': 'foo',
    'body': 'bar',
    'userId': 1
}   
post_response = requests.post('https://jsonplaceholder.typicode.com/posts', json=post_data)
print("POST request to 'https://jsonplaceholder.typicode.com/posts' made successfully.")
print("Status code of the POST response:", post_response.status_code)
print("Content of the POST response:")
print(post_response.text)
# Making a PUT request to a URL
put_data = {
    'id': 1,
    'title': 'foo',
    'body': 'bar',
    'userId': 1
}
put_response = requests.put('https://jsonplaceholder.typicode.com/posts/1', json=put_data)
print("PUT request to 'https://jsonplaceholder.typicode.com/posts/1' made successfully.")
print("Status code of the PUT response:", put_response.status_code)
print("Content of the PUT response:")
print(put_response.text) 
# Making a DELETE request to a URL
delete_response = requests.delete('https://jsonplaceholder.typicode.com/posts/1')
print("DELETE request to 'https://jsonplaceholder.typicode.com/posts/1' made successfully.")
print("Status code of the DELETE response:", delete_response.status_code)
print("Content of the DELETE response:")
print(delete_response.text)
print("All Requests operations executed successfully.")
```
